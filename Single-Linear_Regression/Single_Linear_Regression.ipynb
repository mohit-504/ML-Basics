```python
#making imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (5.0, 5.0)

data = pd.read_csv('data.csv')
X = data.iloc[:,0]
Y = data.iloc[:,1]
print(type(X))
print(Y)

plt.scatter(X,Y)
plt.show()
```

```python
#building model

m=0
c=0
learning_rate=0.0001
epochs = 1000
n = len(X)

for i in range(1000):
    Y_predicted = m * X + c
    D_m = -(2/n) * sum(X * (Y - Y_predicted))
    D_c = -(2/n) * sum(Y - Y_predicted)

    m = m - learning_rate * D_m
    c = c - learning_rate * D_c

plt.plot(X,Y_predicted, color='red')
plt.scatter(X, Y)
plt.show()
print(m,c)
```

```python
#making predictions

input_X = float(input("Enter X coordinate: "))

output_Y = m * input_X + c

print(input_X, output_Y)
plt.plot(X,Y_predicted, color='red')
plt.plot(input_X, output_Y, marker="o", markersize=10, markeredgecolor="red", markerfacecolor="green")
plt.scatter(X, Y)
plt.show()
```
